### Levenshtein Distance

```python
from fuzzywuzzy import fuzz

def similarity(lyric: str, term: str):
    return fuzz.partial_ratio(lyric.lower(), term.lower())
```

```python
df_levenshtein = df_lyrics.copy()
df_levenshtein['levenshtein_distance'] = df_levenshtein['musica_nome'].apply(lambda x: similarity(x, search_term))
df_levenshtein.sort_values(by='levenshtein_distance', ascending=False, inplace=True)
df_levenshtein.head(10)

graph_title = "Melhores resultados para busca: " + search_term + "\""

fig = px.bar(df_levenshtein.head(20),
             x='musica_nome',
             y='levenshtein_distance',
             title="Letras mais próximas de: \"" + search_term + "\"",
             labels={'musica_nome': 'Música', 'levenshtein_distance': 'Distância'},
             color='levenshtein_distance',
             hover_name='cantor_nome')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

### Jaccard

```python
def jaccard_similarity(term: str, lyrics: str):
    search_set = set(term.lower().split())
    lyrics_set = set(lyrics.lower().split())
    intersection = len(lyrics_set.intersection(search_set))
    union = len(lyrics_set.union(search_set))
    return float(intersection / union)
```

```python
df_jaccard = df_lyrics.copy()
df_jaccard['jaccard_similarity'] = df_jaccard['musica_nome'].apply(lambda x: jaccard_similarity(x, search_term))
df_jaccard.sort_values(by='jaccard_similarity', ascending=False, inplace=True)
df_jaccard.head(10)

graph_title = "Melhores resultados para busca: " + search_term + "\""

fig = px.bar(df_jaccard.head(20),
             x='musica_nome',
             y='jaccard_similarity',
             title="Letras mais próximas de: \"" + search_term + "\"",
             labels={'musica_nome': 'Música', 'jaccard_similarity': 'Similaridade'},
             color='jaccard_similarity',
             hover_name='cantor_nome')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

#### Conclusão

Similaridade de Jaccard define o quão próximos dois conjuntos de dados são. Embora ao buscar uma sentença maior, o resultado pareça ser de melhor qualidade, imagino que stop words, caracteres não tratados possam atrapalhar o resultado. Além disso, a comparação entre conjuntos de dados de tamanhos tão distintos parece ser algo a ser ponderado.

### Hamming

```python
def hamming_distance(x, y):
  lmax = max(len(x), len(y))
  x, y = x.rjust(lmax), y.rjust(lmax)
  return sum([a != b for a, b in zip(x, y)])
```

```python
# search_term = input('Buscar por...')

df_hamming = df_lyrics.copy()
df_hamming['hamming_distance'] = df_hamming['musica_nome'].apply(lambda x: hamming_distance(x, search_term))
df_hamming.sort_values(by='hamming_distance', ascending=True, inplace=True)
df_hamming.head(10)

graph_title = "Melhores resultados para busca: " + search_term + "\""

fig = px.bar(df_hamming.head(20),
             x='musica_nome',
             y='hamming_distance',
             title="Letras mais próximas de: \"" + search_term + "\"",
             labels={'musica_nome': 'Música', 'hamming_distance': 'Similaridade'},
             color='hamming_distance',
             hover_name='cantor_nome')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

#### Conclusão

Hamming não parece ser um bom método para realizar busca num texto longo, como a letra de uma música. Ao comparar arrays de mesmo tamanho, caso o termo buscado seja de tamanho muito inferior, o algoritmo compensará a diferença de tamanho com espaços em branco, aumentando a distância entre os valores comparados. O não tratamento de stop words **parece** ser um fator a aumentar a distância.

### Distância do Cosseno

```python
import math
import string
from collections import Counter

# Lista de stop words em português
stop_words = set(
    [
        "o", "a", "os", "as", "de", "do", "da", "dos", "das", "e", "em",
        "no", "na", "nos", "nas", "ao", "à", "às", "pelo", "pela", "pelos", "pelas",
    ]
)


# Função para tokenizar o texto
def tokenize(text, use_bigrams=False):
    text = text.lower().translate(str.maketrans("", "", string.punctuation))
    words = [word for word in text.split() if word not in stop_words]
    if use_bigrams:
        bigrams = [" ".join([words[i], words[i + 1]]) for i in range(len(words) - 1)]
        words.extend(bigrams)
    return words

# Função para calcular a frequência dos termos
def compute_tf(word_dict, words):
    tf_dict = {}
    total_words = len(words)
    for word, count in word_dict.items():
        tf_dict[word] = count / total_words
    return tf_dict

# Função para calcular a frequência inversa dos documentos
def compute_idf(documents):
    N = len(documents)
    idf_dict = {}
    all_words = set(word for doc in documents for word in doc)
    idf_dict = dict.fromkeys(all_words, 0)

    for document in documents:
        for word in set(document):
            idf_dict[word] += 1

    for word, val in idf_dict.items():
        idf_dict[word] = math.log((N + 1) / (float(val) + 1)) + 1

    return idf_dict

# Função para calcular o TF-IDF de cada palavra
def compute_tf_idf(tf, idf):
    tf_idf = {}
    for word, val in tf.items():
        tf_idf[word] = val * idf[word]
    return tf_idf

# Função para calcular a similaridade cosseno entre dois vetores
def cosine_distance(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])

    sum1 = sum([vec1[x] ** 2 for x in vec1.keys()])
    sum2 = sum([vec2[x] ** 2 for x in vec2.keys()])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator

# Função principal para calcular a distância do cosseno entre dois textos
def calcular_distancia_cosseno(texto1, texto2, use_bigrams=False):
    words1 = tokenize(texto1, use_bigrams=use_bigrams)
    words2 = tokenize(texto2, use_bigrams=use_bigrams)

    word_dict1 = Counter(words1)
    word_dict2 = Counter(words2)

    tf1 = compute_tf(word_dict1, words1)
    tf2 = compute_tf(word_dict2, words2)

    idf = compute_idf([words1, words2])

    tf_idf1 = compute_tf_idf(tf1, idf)
    tf_idf2 = compute_tf_idf(tf2, idf)

    return cosine_distance(tf_idf1, tf_idf2)



# Texto a ser comparado com cada letra
texto1 = input("Insira o texto para comparar: ")

# Calcular a distância de cada letra no DataFrame
df_lyrics['distancia_cosseno'] = df_lyrics['musica_nome'].apply(lambda letra: calcular_distancia_cosseno(texto1, letra, use_bigrams=True))

# Exibir as letras ordenadas pela distância
df_lyrics_sorted = df_lyrics.sort_values(by='distancia_cosseno', ascending=False)

df_lyrics_sorted.head(10)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer


def calcular_cosseno(texto1, textos):
    # Ajusta o vetorizador em todo o DataFrame (incluindo o texto de entrada)
    vectorizer = TfidfVectorizer().fit(textos + [texto1])
    tfidf_matrix = vectorizer.transform([texto1] + textos)

    # Calcula a semelhança de cosseno entre o texto de entrada e cada texto no DataFrame
    distances = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()
    return distances


texto1 = input("Insira o texto para comparar: ")


df_lyrics["distancia_cosseno"] = calcular_cosseno(
    texto1, df_lyrics["musica_nome"].tolist()
)


df_lyrics_sorted = df_lyrics.sort_values(by="distancia_cosseno", ascending=False)

df_lyrics_sorted.head(10)
```
